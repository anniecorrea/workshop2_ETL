```python
plt.figure(figsize=(8,5))
sns.histplot(df_spotify['popularity'], bins=30, kde=True)
plt.title('Distribution of Song Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```

```python
df_spotify['duration_min'] = df_spotify['duration_ms'] / 60000

plt.figure(figsize=(8,5))
sns.histplot(df_spotify['duration_min'], bins=30, kde=True)
plt.title('Distribution of Song Duration (minutes)')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Top 10 artists by song count
top_artists = df_spotify['artists'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_artists.values, y=top_artists.index, color="skyblue")
plt.title('Top 10 Artists by Number of Songs')
plt.xlabel('Number of Songs')
plt.ylabel('Artist')
plt.show()
```

```python
top_songs = df_spotify['track_name'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_songs.values, y=top_songs.index, color="pink")
plt.title('Top 10 Songs by Appearance')
plt.xlabel('Count')
plt.ylabel('Song')
plt.show()
```

Grammys Visualizations


```python
# Awards by year
awards_by_year = df_grammys['year'].value_counts().sort_index()
plt.figure(figsize=(12,5))
sns.lineplot(x=awards_by_year.index, y=awards_by_year.values)
plt.title('Number of Grammy Awards by Year')
plt.xlabel('Year')
plt.ylabel('Number of Awards')
plt.show()
```

```python
# Top 10 Grammy categories
top_categories = df_grammys['category'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_categories.values, y=top_categories.index, color ="lightgreen")
plt.title('Top 10 Grammy Award Categories')
plt.xlabel('Number of Awards')
plt.ylabel('Category')
plt.show()
```

```python
# Top 10 Grammy-winning artists
top_grammy_artists = df_grammys['artist'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_grammy_artists.values, y=top_grammy_artists.index, color='violet')
plt.title('Top 10 Grammy-Winning Artists')
plt.xlabel('Number of Awards')
plt.ylabel('Artist')
plt.show()
```

```python
faltantes_por_categoria = (
    df_grammys[df_grammys['artist'].isna() | (df_grammys['artist'] == 'unknown')]
    .groupby('category')
    .size()
    .reset_index(name='artistas_faltantes')
    .sort_values(by='artistas_faltantes', ascending=False)
)

print(faltantes_por_categoria.head(30))
```
